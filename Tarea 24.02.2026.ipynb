Partiendo desde una lista

```python
lista = [1, 2, 3, 4]
print(lista)
```

Lista a Tupla

```python
tupla = tuple(lista)
print(tupla)
```

Lista a Diccionario

```python
diccionario = dict(enumerate(lista))
print(diccionario)
```

Lista a Conjunto

```python
conjunto = set(lista)
print(conjunto)
```

Partiendo desde una tupla

```python
tupla = (1, 2, 3, 4)
print(tupla)
```

Tupla a Lista

```python
lista = list(tupla)
print(lista)
```

Tupla a Diccionario

```python
diccionario = dict(enumerate(tupla))
print(diccionario)
```

Tupla a Conjunto

```python
conjunto = set(tupla)
print(conjunto)
```

Partiendo desde un diccionario

```python
diccionario = {0: 1, 1: 2, 2: 3, 3: 4}
print(diccionario)
```

Diccionario a Lista

```python
lista = list(diccionario)
print(lista)
```

Diccionario a Tupla

```python
tupla = tuple(diccionario)
print(tupla)
```

Diccionario a Conjunto

```python
conjunto = set(diccionario)
print(conjunto)
```

Partiendo desde un conjunto

```python
conjunto = {1, 2, 3, 4}
print(conjunto)
```

Conjunto a Lista

```python
lista = list(conjunto)
print(lista)
```

Conjunto a Tupla

```python
tupla = tuple(conjunto)
print(tupla)
```

Conjunto a Diccionario

```python
diccionario = dict(enumerate(conjunto))
print(diccionario)
```
